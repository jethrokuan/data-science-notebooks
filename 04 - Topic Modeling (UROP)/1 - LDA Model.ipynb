# Loading the Dictionary and Corpus

```python
from gensim.corpora import Dictionary
```

```python
dct = Dictionary.load(f'{PATH}/dictionary.pkl')
```

```python
len(dct)
```

```python
from gensim.corpora import MmCorpus
```

```python
corpus = MmCorpus(f'{PATH}/corpus.mm')
```

# Training the Model

```python
from gensim.models import LdaModel
```

```python
model = LdaModel(corpus=corpus, id2word=dct, num_topics=100, chunksize=500, iterations=1000, alpha='auto')
```
